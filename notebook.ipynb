## Visualization & Explanatory Analysis

### Pertanyaan 1:
### Insight: Weather Conditions vs. Total Bike Rentals 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
# Pemetaan kondisi cuaca ke label yang mudah dibaca
weather_desc = {
    1: 'Clear/Few Clouds',  
    2: 'Mist/Cloudy',
    3: 'Rain/Snow',
    4: 'Thunderstorm'
}

#Terapkan pemetaan langsung ke kolom 'weathersit'
df_all['weathersit'] = df_all['weathersit'].map(weather_desc)
# Kelompokkan berdasarkan kolom 'weathersit' yang baru dan agregasikan jumlah sepeda
df_all_weather = df_all.groupby("weathersit")["cnt"].sum().reset_index()

#Ganti nama kolom 'cnt' menjadi 'total_bike_count'.
df_all_weather.rename(columns={
    "cnt": "total_bike_count"
}, inplace=True)

# Buatlah grafik garis untuk menunjukkan penyewaan sepeda berdasarkan cuaca.
plt.figure(figsize=(10, 6))
sns.barplot(x='weathersit', y='total_bike_count', data=df_all_weather)

# Memformat sumbu y untuk menampilkan angka tanpa notasi ilmiah (1e6 -> 1.000.000).
formatter = FuncFormatter(lambda x, pos: '{:,.0f}'.format(x))  # Menambahkan koma untuk ribuan
plt.gca().yaxis.set_major_formatter(formatter)

plt.title('Bike Rentals by Season')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Rentals')
plt.xticks(rotation=45)
plt.show()

```

##### Berdasarkan analisis barplot, terlihat bahwa  jumlah penyewa sepeda naik pada cuaca cerah/sedikit awan (Clear/Few Clouds), diikuti dengan cuacat kabut/berawan (Mist/Cloudy) , hujan/salju (Rain/Snow) dan terkahir cuaca hujan badai petir(Thunderstorm)

### Pertanyaan 2:
## Season vs. Total Bike Rentals 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

# Pemetaan kondisi musim ke label yang mudah dibaca
season_desc = {
    1: 'Springer',  
    2: 'Summer',
    3: 'Fall',
    4: 'Winter'
}

# Terapkan pemetaan langsung ke kolom 'season'
df_all['season'] = df_all['season'].map(season_desc)

# Kelompokkan berdasarkan kolom 'season' yang baru dan agregasikan jumlah sepeda
df_all_season = df_all.groupby("season")["cnt"].sum().reset_index()

#Ganti nama kolom 'cnt' menjadi 'total_bike_count'.
df_all_season.rename(columns={
    "cnt": "total_bike_count"
}, inplace=True)

# Buatlah grafik garis untuk menunjukkan penyewaan sepeda berdasarkan musim.
plt.figure(figsize=(10, 6))
sns.lineplot(x='season', y='total_bike_count', data=df_all_season)

# Memformat sumbu y untuk menampilkan angka tanpa notasi ilmiah (1e6 -> 1.000.000).
formatter = FuncFormatter(lambda x, pos: '{:,.0f}'.format(x))  # Menambahkan koma untuk ribuan
plt.gca().yaxis.set_major_formatter(formatter)

plt.title('Bike Rentals by Season')
plt.xlabel('Season')
plt.ylabel('Number of Rentals')
plt.xticks(rotation=45)
plt.show()
```

##### Berdasarkan analisis lineplot, terlihat bahwa pada musim gugur (Fall), jumlah sepeda yang disewa lebih banyak dibandingkan musim lainnya, diikuti musim panas (Summer), musim dingin (Winter), dan musim semi (Springer)
