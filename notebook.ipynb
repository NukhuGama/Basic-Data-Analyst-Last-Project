## Visualization & Explanatory Analysis

### Pertanyaan 1:
### Insight: Weather Conditions vs. Total Bike Rentals 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Pemetaan kondisi cuaca ke label yang mudah dibaca
weather_desc = {
    1: 'Clear/Few Clouds',  
    2: 'Mist/Cloudy',
    3: 'Rain/Snow',
    4: 'Thunderstorm'
}

#Terapkan pemetaan langsung ke kolom 'weathersit'
df_all['weathersit'] = df_all['weathersit'].map(weather_desc)

# Buat plot batang untuk menunjukkan distribusi penyewaan sepeda berdasarkan kondisi cuaca
plt.figure(figsize=(10, 6))
sns.barplot(x='weathersit', y='cnt', data=df_all)
plt.title('Bike Rentals by Weather Condition using Merged Dataset')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Rentals')
plt.xticks(rotation=45)
plt.show()
```

### Pertanyaan 2:
## Season vs. Total Bike Rentals 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Pemetaan kondisi musim ke label yang mudah dibaca
season_desc = {
    1: 'Springer',  
    2: 'Summer',
    3: 'Fall',
    4: 'Winter'
}

# Terapkan pemetaan langsung ke kolom 'season'
df_all['season'] = df_all['season'].map(season_desc)

# Buat plot garis untuk menunjukkan distribusi penyewaan sepeda berdasarkan kondisi musim
plt.figure(figsize=(10, 6))
sns.lineplot(x='season', y='cnt', data=df_all)
plt.title('Bike Rentals by Weather Condition using Merged Dataset')
plt.xlabel('Season')
plt.ylabel('Number of Rentals')
plt.xticks(rotation=45)
plt.show()
```
